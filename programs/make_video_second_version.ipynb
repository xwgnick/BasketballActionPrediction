```python
from matplotlib.patches import Circle, Rectangle, Arc
def draw_court(ax=None, color='black', lw=2, outer_lines=False):
    # If an axes object isn't provided to plot onto, just get current one
    if ax is None:
        ax = plt.gca()

    # Create the various parts of an NBA basketball court

    # Create the basketball hoop
    # Diameter of a hoop is 18" so it has a radius of 9", which is a value
    # 7.5 in our coordinate system
    hoop_down = Circle((25, 4.75), radius=0.75, linewidth=lw, color=color, fill=False)
    hoop_up = Circle((25, 89.25), radius=0.75, linewidth=lw, color=color, fill=False)
    # Create backboard
    backboard_down = Rectangle((23, 4), 4, 0, linewidth=lw, color=color)
    backboard_up = Rectangle((23, 90), 4, 0, linewidth=lw, color=color)
    #Create lines
    lines_down = Rectangle((0, 0), 50, 47, linewidth=lw, color=color, fill = False)
    lines_up = Rectangle((0, 47), 50, 47, linewidth=lw, color=color, fill = False)
    # The paint
    # Create the outer box 0f the paint, width=16ft, height=19ft
    painted_down = Rectangle((19,0), 12, 18.83, linewidth=lw, color=color,
                          fill=False)
    painted_up = Rectangle((19, 75.1), 12, 18.83, linewidth=lw, color=color,
                          fill=False)
    # Create the inner box of the paint, widt=12ft, height=19ft
#     inner_box_left = Rectangle((0, 95), 101, 60, linewidth=lw, color=color,
#                           fill=False)
#     inner_box_right = Rectangle((367, 95), 101, 60, linewidth=lw, color=color,
#                           fill=False)
    # Create free throw top arc
    free_throw_down = Arc((25, 18.83), 12, 12, theta1= 0, theta2 = 180,
                         linewidth=lw, color=color, fill=False)
    free_throw_up = Arc((25, 75.1), 12, 12, theta1=-180, theta2=0,
                         linewidth=lw, color=color, fill=False)
    # Create free throw bottom arc
#     bottom_free_throw_left = Arc((101, 125), 60, 60, theta1=90, theta2=270,
#                             linewidth=lw, color=color, linestyle='dashed')
#     bottom_free_throw_right = Arc((367, 125), 60, 60, theta1=-90, theta2=90,
#                             linewidth=lw, color=color, linestyle='dashed')
    # Restricted Zone, it is an arc with 4ft radius from center of the hoop
#     restricted = Arc((0, 0), 80, 80, theta1=0, theta2=180, linewidth=lw,
#                      color=color)

    # Three point line
    # Create the side 3pt lines, they are 14ft long before they begin to arc
    corner_three_a_down = Rectangle((4.25,0), 0, 5, linewidth=lw, color=color)
    corner_three_b_down = Rectangle((45.75, 0), 0, 5, linewidth=lw, color=color)
    # 3pt arc - center of arc will be the hoop, arc is 23'9" away from hoop
    # I just played around with the theta values until they lined up with the 
    # threes
    three_arc_down = Arc((25, 4.25), 41.5, 41.5, theta1=0, theta2=180, linewidth=lw, color=color)
    corner_three_a_up = Rectangle((4.25,89), 0, 5, linewidth=lw, color=color)
    corner_three_b_up = Rectangle((45.75, 89), 0, 5, linewidth=lw, color=color)
    # 3pt arc - center of arc will be the hoop, arc is 23'9" away from hoop
    # I just played around with the theta values until they lined up with the 
    # threes
    three_arc_up = Arc((25, 89.25), 41.5, 40.7, theta1=-180, theta2=0, linewidth=lw,
                    color=color)
    # Center Court
    center_inner_circle = Circle((25, 47), radius=6, linewidth=lw, color=color, fill=False)
#     center_outer_circle = Circle((234, 125), radius=32, linewidth=lw, color=color, fill=False)
    
    court_elements = [hoop_down, hoop_up, backboard_down, backboard_up ,lines_down, lines_up, painted_down, painted_up, free_throw_down, free_throw_up, corner_three_a_down, corner_three_b_down, three_arc_down, corner_three_a_up,corner_three_b_up, three_arc_up, center_inner_circle]
#     court_elements = [hoop_left,hoop_right,backboard_left, backboard_right,lines_left, lines_right,outer_box_left,outer_box_right,inner_box_left
#                      ,inner_box_right,top_free_throw_left,bottom_free_throw_left,top_free_throw_right,bottom_free_throw_right,corner_three_a_left, 
#                       corner_three_b_left,three_arc_left,corner_three_a_right,corner_three_b_right,three_arc_right,
#                      center_inner_circle,center_outer_circle]

    for element in court_elements:
        ax.add_patch(element)

    return ax
```

```python
def init(): 
    ax.add_patch(ball_holder)
    ax.add_patch(ball)
    i = 0
    while i < len(player_circles):
        ax.add_patch(player_circles[i])
        i += 1
    return tuple(player_text) + tuple(player_circles) + (ball_holder,) + (ball, )

def animate(i):
    j = 0
#     print(d["ball_x"].iloc[0])
    ball_x = d["ball_x"].iloc[i]
    ball_y = d["ball_y"].iloc[i]
    ball_holder_x_name = "p" + str(d["p.poss"].iloc[i] + 1) + "_x"
    ball_holder_y_name = "p" + str(d["p.poss"].iloc[i] + 1) + "_y"
    ball_holder_x = d[ball_holder_x_name].iloc[i]
    ball_holder_y = d[ball_holder_y_name].iloc[i]
    title_line = "Frame " + str(i)
    title.set_text(title_line)
    number_move = reverse_dict(d_class.real_label_number)
    if d["home"].iloc[i] == h_v:
        sample_frame = d_class.nn_ready.loc[d.index.values[i]]
        ps = loaded_prediction("../../saved_models/model_crossvalidated", sample_frame, reverse_dict(d_class.real_label_number), mode = "one", verbose = False)
        k = 0
        for prob in ps:
            text_line = "There are " + str(round(prob,2) * 100) + "% probability that the ball-holder's next move is " + number_move[k]
            prediction_line_list[k].set_text(text_line)
            k += 1
        prediction_line = "The prediction for the ball holder's next move is: " + number_move[ps.index(max(ps))]
        prediction_line_list[7].set_text(prediction_line)
    else:
        prediction_line = "No prediction for non-Duke players"
        o = 0
        while o < 7:
            prediction_line_list[o].set_text("")
            o += 1
        prediction_line_list[7].set_text(prediction_line)
    current_move = "Ballholder current move is: " + number_move[d["event.descrip"].iloc[i]]
    current_move_line.set_text(current_move)

    while j < len(player_circles):
        column_name_x = "p" + str(j + 1) + "_x"
        column_name_y = "p" + str(j + 1) + "_y"
        player_circles[j].center = (d[column_name_x].iloc[i], d[column_name_y].iloc[i])
        player_text[j].set_text(str(j + 1))
        player_text[j].set_x(d[column_name_x].iloc[i])
        player_text[j].set_y(d[column_name_y].iloc[i])
        j += 1
    ball_holder.center = (ball_holder_x, ball_holder_y)
    ball.center = (ball_x, ball_y)
    return tuple(player_text) + tuple(player_circles) + (ball_holder,) + (ball, )
```

```python
def paint_one_frame(other_attribute, position_list, i, title):
    j = 0
    fig = plt.figure(figsize=(12,6))
    ax = plt.gca(xlim = (0 ,467),ylim = (0, 250)) #create axis object
    draw_court()
    title_line = "Frame " + str(i)
    title.set_text(title_line)
    ball_x = position_list[-1][i][0] * 5
    ball_y = position_list[-1][i][1] * 5
    shortest_dis = 200000
    ballholding_player = None
    for player in position_list:
        player_x = position_list[player][i][0] * 5
        player_y = position_list[player][i][1] * 5
        if player in jersey_pid_visitor:
            player_circle = plt.Circle((player_x, player_y), 5, color="blue")
        elif player == -1:
            player_circle = plt.Circle((player_x, player_y), 2, color="orange")
        else:
            player_circle = plt.Circle((player_x, player_y), 5, color="red")
        ax.add_patch(player_circle)
        if player is not -1:
            player_text = ax.text(player_x, player_y,str(jersey_pid[player]),color="white",ha='center',va='center')
            ax.add_patch(player_circle)
            ball_dis = math.sqrt((ball_x - player_x) **2 + (ball_y - player_y) **2)
            if(ball_dis < shortest_dis):
                shortest_dis = ball_dis
                ballholding_player = player
    ball_holder_x = position_list[ballholding_player][i][0] * 5
    ball_holder_y = position_list[ballholding_player][i][1] * 5
    x_axis = Rectangle((0,ball_holder_y), 446, 0, linewidth=1, color="black")
    y_axis = Rectangle((ball_holder_x,0), 0, 250, linewidth=1, color="black")
    ax.add_patch(x_axis)
    ax.add_patch(y_axis)
    ballholder_team = None
    if ballholding_player in jersey_pid_visitor:
        ballholder_team = "visitor"
    else:
        ballholder_team = "home"
        
    player_coords = {}
    iti = 0
    for player in position_list:
        player_team = None
        same_team = 0
        if player in jersey_pid_visitor:
            player_team = "visitor"
        else:
            player_team = "home"
        if player_team == ballholder_team:
            same_team = 1
        pair = []
        player_x = position_list[player][i][0] * 5
        player_y = position_list[player][i][1] * 5
        relative_x = round(player_x - ball_holder_x,1)
        relative_y = round(player_y - ball_holder_y,1)
        rela_line = "(" + str(relative_x) + "," + str(relative_y) + ")"
        player_relative_text = ax.text(player_x, player_y + 10,rela_line,color="black",ha='center',va='center')
#         ax.add_patch(player_relative_text)
        if (player == ballholding_player) or (player == -1):
            continue
        pair.append(relative_x)
        pair.append(relative_y)
        pair.append(player_x)
        pair.append(player_y)
        pair.append(same_team)
        player_coords[player] = pair
        iti += 1
    player_coords["game_clock"] = other_attribute["game_clock"][i]
    player_coords["shot_clock"] = other_attribute["shot_clock"][i]
    plt.show()
    return player_coords
#         player_rela_coo_list.append(player_rela_coords)
        
#         ballholder_list.append(ballholding_player)
```
